```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from matplotlib.pyplot import style
```

```python
Link = r'https://drive.google.com/file/d/1Al8t5j84VkisJOLzckfnxDmEKK7muT6p/view?usp=sharing'
pd_link='https://drive.google.com/uc?id='+ Link.split('/')[-2]
df = pd.read_csv(pd_link)
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
df_hd=df.head(50)
df_hd
```

```python
per = df_hd.isnull().sum()/df_hd.shape[0] *100
per 
```

```python
error_key = per[per >35].keys()
error_key
```

```python
df2 =df_hd.drop(columns= error_key)
df2
```

```python
plt.figure(figsize =(12,5))
sns.heatmap(df2.isnull())
plt.show()
```

```python
df_int = df2.select_dtypes(include = ['int64','float64'])
df_int
```

```python
df_int[df_int.isnull().any(axis = 1) ]
```

```python
df_int.isnull().sum()
```

```python
list1 = df_int.columns
list1
```

```python
lt = []
for i in range (len(list1)):
    if  df_int[list1[i]].isnull().sum() >0 :
        lt.append(list1[i])
        
```

```python
lt
```

```python
plt.figure(figsize =(20,15))
for i in range (len(lt)):
    plt.subplot(4,3,i+1)
    sns.histplot(df_int[lt[i]],color="g", label="100% Equities", kde=True, stat="density", linewidth=0.5)
#     sns.histplot(a, color="red", label="100% Equities", kde=True, stat="density", linewidth=0)
```

```python
mean_value = df_int.mean(numeric_only=True) 
mean_value
```

```python
df3 = df_int.fillna(df_int.mean(numeric_only=True) )
df3
```

```python

plt.figure(figsize =(12,5))
sns.heatmap(df3.isnull())
plt.show()
```

```python
df_org = df2
df_clean_mean = df3
```

```python
plt.figure(figsize =(20,15))
for i in range (len(lt)):
    plt.subplot(4,3,i+1)
    sns.histplot( df_org[lt[i]]   ,color="g", label="100% Equities", kde=True, stat="density", linewidth=0.5)
    sns.histplot( df_clean_mean[lt[i]]   ,color="y", label="100% Equities", kde=True, stat="density", linewidth=0.5)
    plt.legend(['Original','CleanData' ],fontsize =12.5)
```

```python
df3_median = df_int.fillna(df_int.median(numeric_only=True) )
df_clean_median =df3_median
df_clean_median
```

```python
plt.figure(figsize =(25,18))
for i in range (len(lt)):
    plt.subplot(4,3,i+1)
    sns.histplot( df_org[lt[i]]   ,color="r", label="100% Equities", kde=True, stat="density", linewidth=0.5)
    sns.histplot( df_clean_mean[lt[i]]   ,color="g", label="100% Equities", kde=True, stat="density", linewidth=0.5)
    sns.histplot( df_clean_median[lt[i]]   ,color="y", label="100% Equities", kde=True, stat="density", linewidth=0.5)
    
    plt.legend(['Original','CleanData_mean','Clean_Data_mean' ],fontsize =12)
```

```python
plt.figure(figsize =(25,18))
for i in range (len(lt)):
    plt.subplot(4,3,i+1)
    plt.boxplot( df_org[lt[i]]   )
    plt.boxplot( df_clean_mean[lt[i]] )
    plt.boxplot( df_clean_median[lt[i]] )
    
    plt.legend(['Original','CleanData_mean','Clean_Data_median' ],fontsize =12)
```

```python
df_concatinate = pd.concat([df_org[lt],df_clean_mean[lt],df_clean_median[lt]],axis= 1)
df_concatinate
```

```python
# after Concatination you could able to see that what value place at the NaN after filling the NaN 
df_concatinate[df_concatinate.isnull().any(axis= 1)]
```
